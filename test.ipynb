```python
import pandas as pd
import numpy as np

import sklearn.datasets


data= sklearn.datasets.load_iris()
```

```python
x=data.data
y=data.target
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
# svm=SVC()
params={
    "C":[0,0.5,1,5,10],
    "kernel":["linear","rbf"],
    "gamma":["auto",0.1],
    "probability":[True,False]
    
}


```

```python
gs=GridSearchCV(SVC(),param_grid=params,cv=3,scoring="accuracy",n_jobs=1,verbose=10)
```

```python
gs.fit(x,y)
```

```python
gs.best_score_
```
