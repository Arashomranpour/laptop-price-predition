```python
df["Memory"]=df["Memory"].astype(str).replace("\.0","",regex=True)
df["Memory"]=df["Memory"].str.replace("GB","")
df["Memory"]=df["Memory"].str.replace("TB","000")
new=df["Memory"].str.split("+",n=1,expand=True)

df["First"]=new[0]
df["First"]=df["First"].str.strip()

df["Sec"]=new[1]

df["Layer1hdd"]=df["First"].apply(lambda x:1 if "HDD" in x else 0)
df["Layer1sdd"]=df["First"].apply(lambda x:1 if "SSD" in x else 0)
df["Layer1Hybrid"]=df["First"].apply(lambda x:1 if "Hybrid" in x else 0)
df["Layer1FlashStorage"]=df["First"].apply(lambda x:1 if "Flash Storage" in x else 0)

df["First"]=df["First"].str.replace(r'\D','')

df["Sec"].fillna("0",inplace=True)

df["Layer2hdd"]=df["Sec"].apply(lambda x:1 if "HDD" in x else 0)
df["Layer2sdd"]=df["Sec"].apply(lambda x:1 if "SSD" in x else 0)
df["Layer2Hybrid"]=df["Sec"].apply(lambda x:1 if "Hybrid" in x else 0)
df["Layer2FlashStorage"]=df["Sec"].apply(lambda x:1 if "Flash Storage" in x else 0)

df["Sec"]=df["Sec"].str.replace(r'\D','')

df['First']=df["First"].astype(int)
df['Sec']=df["Sec"].astype(int)


df["HDD"]=(df["First"]*df["Layer1hdd"]+df["Sec"]*df["Layer2hdd"])
df["SDD"]=(df["First"]*df["Layer1sdd"]+df["Sec"]*df["Layer2sdd"])
df["Hybrid"]=(df["First"]*df["Layer1Hybrid"]+df["Sec"]*df["Layer2Hybrid"])
df["FlashStorage"]=(df["First"]*df["Layer1FlashStorage"]+df["Sec"]*df["Layer2FlashStorage"])
```

```python
from sklearn.ensemble import VotingRegressor,StackingRegressor

step1=ColumnTransformer(transformers=[('Col_tr',OneHotEncoder(sparse=False,drop="first"),[0,1,3,8,11,12])],remainder="passthrough")

rf=RandomForestRegressor(n_estimators=100,random_state=3,max_depth=15,max_features=0.75,max_samples=0.5)
xrbos=GradientBoostingRegressor(n_estimators=500)
# xgb=XGBRegressor(gamma=0,learning_rate=0.5,max_depth=4)
et=ExtraTreesRegressor(n_estimators=100,random_state=3,max_depth=15,max_features=0.75)

step2=VotingRegressor([("rf",rf),("grbos",xrbos),("et",et)])

pipe=Pipeline([('step1',step1),("step2",step2)])
pipe.fit(X_Train,Y_Train)
ypred=pipe.predict(X_Test)

print(r2_score(Y_Test,ypred))
print(mean_absolute_error(Y_Test,ypred))

import pickle
pickle.dump(df,open("df.pkl","wb"))
pickle.dump(pipe,open("pipe1.pkl","wb"))
```
